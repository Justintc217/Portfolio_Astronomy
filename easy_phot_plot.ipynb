```python
from astropy.io import ascii
data = ascii.read("lc.dat")  
print(data)
```

```python
x = data["time"]
y1 = data["mag1"]
y2 = data["mag2"]
y3 = data["mag3"]
y4 = data["mag4"]
y5 = data["mag5"]
y_ave = (y2 + y3 + y4 + y5)/4

y_target_ave_dif = y1 - y_ave
print(len(y1) , len(y_ave) , len(y_target_ave_dif))
```

```python
fig, ax = plt.subplots(3,1)
fig.set_size_inches(12,12)

ax[0].scatter(x , y1 , c = "red")
ax[1].scatter(x , y_ave , c = "white")
ax[2].scatter(x , y_target_ave_dif , c = "black" , marker = "*")
"""ax.scatter(x , y2 , c = "orange")
ax.scatter(x , y3 , c = "green")
ax.scatter(x , y4 , c = "blue")
ax.scatter(x , y5 , c = "purple")"""
plt.show()
```
