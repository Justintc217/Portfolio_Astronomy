```python
z_col = np.array([])
element_col = np.array([])
system_col = np.array([])
system_col_nc = np.array([])
for i in [128 , 110 , 103 , 100]:
    orph_form = float("{0:.5f}".format(orph_opts[i]))
    z_col = np.append(z_col , orph_form)
    dummy = str(lam_list[i]) + " " + name_list[i]
    element_col = np.append(element_col , dummy)
    system_col = np.append(system_col , "")
    system_col_nc = np.append(system_col_nc , "")

save_HIz = np.array([])
save_HIe = np.array([])
save_HIs = np.array([])
for i in range(0 , len(large_systems)):
    cond = np.isclose(large_systems[i] , orph_opts , rtol = 0.005) # increase rtol to get more results, will be offset however

    for p in range(0 , len(cond)):
        if cond[p] == True:
            dummy = str(lam[p]) + " " + namer[p]
            orph_form = "{0:.5f}".format(orph_opts[p])
            
            if namer[p] == "HI":
                save_HIz = np.append(save_HIz , orph_form)
                save_HIe = np.append(save_HIe , dummy)
                save_HIs = np.append(save_HIs , large_systems[i])
            
            else:
                z_col = np.append(z_col , orph_form)
                element_col = np.append(element_col , dummy)
                system_col = np.append(system_col , large_systems[i])
                system_col_nc = np.append(system_col_nc , large_systems_nc[i])

delta = np.array([])
for i in range(0 , len(z_col)):
    final = float(z_col[i])
    if system_col[i] == "":
        initial = 0
    else:
        initial = float(system_col[i])
    form_fi = "{0:.6f}".format(final - initial)
    rank = abs(float(form_fi))
    if rank > 0.001:
        form_fi = form_fi + " awful"
    elif rank > 0.00065 and rank <= 0.001:   # above 200km offset
        form_fi = form_fi + "   bad"
    elif rank > 0.0005 and rank <= 0.00065:   # above 150 km offset
        form_fi = form_fi + "  poor"
    elif rank > 0.0003 and rank <= 0.0005:    # above 90 km offset
        form_fi = form_fi + " maybe"
    elif rank > 0.0001 and rank <= 0.0003:    # above 30 km offset
        form_fi = form_fi + "  good"
    else: 
        form_fi = form_fi + " great"
    delta = np.append(delta , form_fi)
```
